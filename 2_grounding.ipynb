Simulating a Hallucination (The Risk)

```python
def demonstrate_hallucination():
    print("=== SIMULATING UNGROUNDED ASSISTANT (High Risk of Hallucination) ===")
    
    # We ask a question about our specific policy WITHOUT giving the model the data
    prompt = "What is the company policy for buying a bottle of wine during an international work trip?"
    
    # Running with higher temperature to show how it creates an answer
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7 
    )
    print(f"User Asked: {prompt}")
    print(f"AI Response:\n{response.choices[0].message.content}\n")
demonstrate_hallucination()
```

Grounding the AI (The Solution)

```python
def demonstrate_grounding():
    print("=== SIMULATING GROUNDED ASSISTANT (Strictly Anchored) ===")
    
    user_question = "What is the company policy for buying a bottle of wine during an international work trip?"
    
    # 1. Give strict behavioral boundaries
    system_instruction = "You are a factual compliance checker. Answer the user question using ONLY the provided context. If the context does not explicitly allow it, state that it is not permitted."
    
    # 2. Glue the private context and the user's question together
    full_prompt = f"Context:\n{company_policy_context}\n\nQuestion: {user_question}"
    
    # Running with temperature 0 for maximum factual consistency
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": system_instruction},
            {"role": "user", "content": full_prompt}
        ],
        temperature=0.0 
    )
    print(f"User Asked: {user_question}")
    print(f"AI Response:\n{response.choices[0].message.content}\n")
    
demonstrate_grounding()
```
